```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Evaluation 1

### 1.1 Import data

```python
df = pd.read_csv('ukf_output1.csv')
df.head()
```

### 1.2 NIS 

```python
t = np.arange(1,df.shape[0]+1,1) # create an array of evenly spaced values (step value)
std = np.array([7.8]*(df.shape[0]))

########### Plotting #################################
plt.figure(figsize=(18,8))
plt.plot(t,df.ix[:,9],'-',t,std,'r--')
plt.ylim(0,20)
plt.xlim(1,500)
plt.title('NIS consistancy check')
plt.xlabel('Step')
plt.ylabel('NIS')
plt.annotate('NIS=7.8', xy=(200, 7.8), xytext=(250, 14),
            arrowprops=dict(facecolor='black', shrink=0.05), )


plt.savefig('ukf_nis1.png') 
```

### 1.3 RMSE

```python
rmse_x = df.ix[:,10]
rmse_y = df.ix[:,11]
rmse_vx = df.ix[:,12]
rmse_vy = df.ix[:,13]
# your px, py, vx, and vy RMSE should be less than or equal to the values [.09, .10, .40, .30].
std_x, std_y, std_vx, std_vy = np.array([.09]*(df.shape[0])), np.array([.10]*(df.shape[0])),\
                            np.array([.40]*(df.shape[0])), np.array([.30]*(df.shape[0]))
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(18, 10))
ax1.plot(t,rmse_x,'-',t,std_x,'r--')
ax1.get_xaxis().set_ticks([]) # remmove the xticks
ax1.annotate('rmse_x=.09', xy=(200, 0.1), xytext=(220, 0.15), arrowprops=dict(facecolor='black', shrink=0.05),)
ax2.plot(t,rmse_y,'-',t,std_y,'r--')
ax2.get_xaxis().set_ticks([]) # remmove the xticks
ax2.annotate('rmse_y=.10', xy=(90, 0.1), xytext=(110, 0.04), arrowprops=dict(facecolor='black', shrink=0.05),)
ax3.plot(t,rmse_vx,'-',t,std_vx,'r--')
ax3.get_xaxis().set_ticks([]) # remmove the xticks
ax3.annotate('rmse_vx=.40', xy=(200, 0.8), xytext=(220,2.0), arrowprops=dict(facecolor='black', shrink=0.05),)
ax4.plot(t,rmse_vy,'-',t,std_vy,'r--')
ax4.annotate('rmse_vy=.30', xy=(300, 0.32), xytext=(320, 0.55), arrowprops=dict(facecolor='black', shrink=0.05),)

plt.savefig('ukf_rmse1.png') 
```
